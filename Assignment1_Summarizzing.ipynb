1. Navigate to Narendra Modi's Wikipedia, and fool around. I cannot legally ask you to do anything else.
2. Write a script that _crawls_ the page, that is, it returns a list of all links on the page. Then for each link, it returns all links and so on. Do this for two levels.
**BONUS** : Write a script that can crawl for N levels, where N is user specified. 

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import os
import wget

import time


n=int(input("Enter "))
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))

link="https://en.wikipedia.org/wiki/Narendra_Modi"

texts=[]


def myfunc(n,link):
    if n==0:
        return
    driver.get(link)
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(2)
        p1=len(texts)
        
        Tweets=[]
        Tweets=list(driver.find_elements(By.XPATH, "//a[@href]" ))
        p=len(Tweets)
        for e in range(p):
            texts.append(Tweets[e].get_attribute("href"))
            print(Tweets[e].get_attribute("href"))
           
        p2=len(texts)  



        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
    for i in range (p1-1,p2):
        myfunc(n-1,texts[i])
    
        
    
        
myfunc(n,link)

return texts
```
